```python
import pandas as pd

def delivery(cart_value):
    if cart_value < 50:
        final = cart_value+4.95
    if cart_value >=50 and cart_value < 90:
        final = cart_value+2.95
    if cart_value >= 90:
        final = cart_value        
    print("Your final total including delivery charges: ", final)
    
def price(cart_items):
    red_count=cart_items.count('RF1')
    green_count=cart_items.count('GF1')
    blue_count=cart_items.count('BF1')
    
    even_count = 0
    for i in range(0,red_count-1):
        if i % 2 == 0:
            even_count += 1
    
    #Assuming offer only for red flowers as given in problem statement
    red_flower = 32.95
    green_flower = 24.95 
    blue_flower = 7.95
    offered_red_price = (red_count*red_flower)-(even_count*(red_flower/2))
    
    cart_items_price = offered_red_price+(green_count*green_flower)+(blue_count*blue_flower)
        
    return delivery(cart_items_price)

def items():
    try:
        cart = []
        count = int(input("Enter number of items wanted to purchase: "))
        for x in range (count):
            val = (input ("Enter each item using product code: ")).upper()
            if val in ['RF1', 'GF1', 'BF1']:
                cart.append(val)
                print(f'Your cart has {cart}')
            else:
                return ('Invalid product code, please start from beginning')
        return price(cart)
    except ValueError:
        print("Entered value is incorrect, please enter number of items ")
        items()

# shows catalog to the customer
catalog = {'PRODUCT': ['Red Flower', 'Green Flower', 'Blue Flower'], 
           ' CODE': ['RF1', 'GF1', 'BF1'], '   PRICE': ['$32.95', '$24.95', '$7.95']}
df = pd.DataFrame(data=catalog)
print("Welcome to Dev Flower Shop Company")
print("Enter product code as mentioned below to add products to your cart\n")
print(df.to_string(index=False), "\n")
print("SPECIAL OFFERS: 'Buy one red flower, Get second one at half price'\n")
print("DELIVERY:")
dff_delivery = {'Total order': ['Below $50', '$50 and Below $90', '$90 and Above'], 'Charges': ['$4.95', '$2.95', '      Free Delivery']}
df_delivery = pd.DataFrame(data=dff_delivery)
print(df_delivery.to_string(index=False),"\n")
items()
```
